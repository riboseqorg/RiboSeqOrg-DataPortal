```python
with open('/home/jack/projects/RiboSeqOrg-DataPortal/data/trips_model.tsv', 'w') as f:
    for row in files.iterrows():
        match = filtered[filtered['Run'] == row[1]['file_name'].split('.')[0]]
        if len(match['Run'].values) == 0:
            run = 'NA'
            bioproject = 'NA'
        else:
            run = match['Run'].values[0]
            bioproject = match['BioProject'].values[0]
        organism = organisms[organisms['organism_id'] == row[1]['organism_id']]
        if organism.shape[0] == 0:
            continue
        study = studies[studies['study_id'] == row[1]['study_id']]
        f.write(f"{bioproject}\t{run}\t{row[1]['file_id']}\t{row[1]['file_name'].split('.')[0]}\t{study['study_name'].values[0]}\t{study['srp_nos'].values[0]}\t{study['gse_nos'].values[0]}\t{study['paper_pmid'].values[0]}\t{organism['organism_name'].values[0]}\t{organism['transcriptome_list'].values[0]}\n")
```

```python
updated_df = pd.read_csv("/home/jack/projects/RiboSeqOrg-DataPortal/data/standardized_columns_final.csv")
df = pd.read_csv("/home/jack/projects/RiboSeqOrg-DataPortal/data/filtered_riboseq_done_260623.csv")
```

```python
clean_rows = []
for _, row in df.iterrows():
    updated_df_row = updated_df[updated_df['Run'] == row['Run']]
    if updated_df_row.shape[0] == 0:
        clean_rows.append(row)
        continue
    if str(updated_df_row['TISSUE_st'].values[0]) not in ['nan', "NA", '']:
        row['TISSUE'] = updated_df_row['TISSUE_st'].values[0]
    elif str(updated_df_row['TISSUE_st'].values[0]) not in ['nan', "NA", '']:
        row['CELL_LINE'] = updated_df_row['CELL_LINE_st'].values[0]
    elif str(updated_df_row['TISSUE_st'].values[0]) not in ['nan', "NA", '']:
        row['INHIBITOR'] = updated_df_row['INHIBITOR_st'].values[0]
    elif str(updated_df_row['TISSUE_st'].values[0]) not in ['nan', "NA", '']:
        row['CONDITION'] = updated_df_row['CONDITION_st'].values[0]
    elif str(updated_df_row['TISSUE_st'].values[0]) not in ['nan', "NA", '']:
        row['REPLICATE'] = updated_df_row['REPLICATE_st'].values[0]
    elif str(updated_df_row['TISSUE_st'].values[0]) not in ['nan', "NA", '']:
        row['LIBRARYTYPE'] = updated_df_row['LIBRARYTYPE_st'].values[0]
    elif str(updated_df_row['TISSUE_st'].values[0]) not in ['nan', "NA", '']:
        row['FRACTION'] = updated_df_row['FRACTION_st'].values[0]
    elif str(updated_df_row['TISSUE_st'].values[0]) not in ['nan', "NA", '']:
        row['TIMEPOINT'] = updated_df_row['TIMEPOINT_st'].values[0]
    # row.drop(['TISSUE_st', 'CELL_LINE_st', 'INHIBITOR_st', 'CONDITION_st', 'REPLICATE_st', 'LIBRARYTYPE_st', 'FRACTION_st', 'TIMEPOINT_st'], inplace=True)
    clean_rows.append(row)

clean_df = pd.DataFrame(clean_rows)
clean_df.head()
```
