# From the Merged Metadata produce the final set in a csv that is to be loaded into DB


```python
import pandas as pd
```

Load in the csv. This is the product of the metadata cleaning scripts. It has x, y and st variants of core columns. 

```python
df = pd.read_csv('../data/Cleaned_Metadata_Merged_w_Manual_x_y_st_metadata_merged_with_manual.csv')
df.head()
```

drop the columns that end with x or y and keep the _st columns

```python
xy_columns = [i for i in df.columns if i.endswith('.x') or i.endswith('.y')]
df.drop(columns=xy_columns, inplace=True)
```

```python
def remove_suffix(column_name, suffix):
    if column_name.endswith(suffix):
        return column_name[:-len(suffix)]
    else:
        return column_name

current_columns = df.columns
df.head()
```

```python
df.drop(columns=["name", 'not_unique'], inplace=True)
```

Write these values into a df with all columns (including those from SRA)

```python
full_df = pd.read_csv('../data/filtered_riboseq_done_260623.csv')
full_df.head()
```

```python
update_columns = [col for col in df.columns if col.endswith("_st")]

for col in update_columns:
    base_col = col[:-3]
    full_df[base_col] = df[col]
full_df.head()
```

write the df to file

```python
full_df.to_csv('../data/Cleaned_Metadata_For_Upload.csv', index=False)
```
